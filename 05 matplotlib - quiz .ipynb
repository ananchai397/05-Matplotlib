```python
#1.สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x=[1,2,3,4,5,6,7,8,9]
x=np.array(x)
y=x^2
plt.plot(x,y)
plt.xlabel('1-9')
plt.ylabel('1-9 squared')
plt.title('quiz1')
plt.show()
```

```python
#2.สร้างกราฟฟังก์ชั่น Sin (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีชมพู 
x=np.arange(0,90)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel('0-90')
plt.ylabel('sin x')
plt.title('quiz2')
plt.plot(x,y,c='#ff00cc')
plt.show()
```

```python
#3.สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีเขียวอ่อนโดยใช้ค่ารหัสสีฐานสิบหก
x=[1,2,3,4,5,6,7,8,9]
x=np.array(x)
y=x^2
plt.plot(x,y)
plt.xlabel('1-9')
plt.ylabel('1-9 squared')
plt.title('quiz3')
plt.plot(x,y,c='#00ff00')
plt.show()
```

```python
#4.สร้างกราฟฟังก์ชั่น Cos (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีแดง ใส่ Line Style เป็นเส้น - - 

x=np.arange(0,90)
y=np.cos(x)
plt.plot(x,y)
plt.xlabel('0-90')
plt.ylabel('cos x')
plt.title('quiz4')
plt.plot(x,y,c='r',ls = '--')

plt.show()
```

```python
#5. สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 และ Y = X^3 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน ใช้สีแดงและน้ำเงินตามลำดับ 
#โดยเส้นแรกใช้วงกลมเป็น marker และเส้นสองใช้สามเหลี่ยมเป็น marker

x=np.arange(0,20)
y1=x^2
y2=x^3
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x-plot')
plt.ylabel('y-plot')
plt.title('quiz5')
plt.plot(x,y1,c='r', marker='o')
plt.plot(x,y2,c='b', marker = '^')
plt.show()
```

```python
#6. สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 และ X = Y^2 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน ใช้สีแดงและน้ำเงินตามลำดับ 
#โดยเส้นแรกใช้วงกลมเป็น marker และเส้นสองใช้สามเหลี่ยมเป็น marker และใส่ legend และเพิ่มเส้นตารางข้างหลัง

x=np.arange(0,20)
y1=x^2
y2=x^3
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x-plot')
plt.ylabel('y-plot')
plt.title('quiz6')
plt.plot(x,y1,c='r', marker='o')
plt.plot(x,y2,c='b', marker = '^')
plt.plot(x,y1,label='x^2')
plt.plot(x,y2,label='x^3')
plt.legend(loc=4)
plt.grid()
plt.show()
```

```python
#7. สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 + 4*Y พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีเหลืองอ่อนโดยใช้ค่ารหัสสีฐานสิบหก
#จำกัดแกน X และ Y ให้เหมาะสม 

y=np.arange(0,20)
x =(y^2)+ (4*y) 
plt.xlabel('x-plot')
plt.ylabel('y-plot')
plt.title('quiz7')
plt.plot(x,y,c='#ffff66')
plt.ylim([0,20])
plt.xlim([0,100])
plt.show()
```

```python
#8. สร้างกราฟใหญ่ด้วยสมการ X = Y^3 และกราฟเล็กด้วย Y = X^2 + 2 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน และใส่ legend 

y=np.arange(0,20)
x=y^3
fig = plt.figure()
ax1=fig.add_axes([0.1,0.1,0.9,0.99])
ax1.plot(x,y)
plt.title('quiz8')
plt.xlabel('x-plot')
plt.ylabel('y-plot')
plt.plot(x,y,label='Y^3')
plt.legend(loc=1)

ax2 = fig.add_axes([0.7,0.17,0.3,0.4])
ax2.plot(y,(x^2)+2)
plt.title('quiz8 mini')
plt.xlabel('x-plot mini')
plt.ylabel('y-plot mini')
```

```python
#9. ให้หาอัตราส่วนตั๋วโดยสารชั้น 1 2 3 และพล็อตเป็น Pie chart 

import pandas as pd
import numpy as np
df = pd.read_csv('train.csv')
df

R=len(df['Pclass'])  #จำนวนทั้งหมด
N=df[(df['Pclass']==1)] 
One =((N['Pclass'].sum())/R)*100

N2=df[(df['Pclass']==2)] 
Two =((N2['Pclass'].sum())/R)*100

N3=df[(df['Pclass']==3)] 
Three =((N3['Pclass'].sum())/R)*100

plt.figure(figsize=[5,5])
labels = ["Pclass 1", "Plass 2", "Pclass 3"]
plt.pie([One,Two,Three], labels=labels)
plt.show()
```

```python
#10. ให้หาอัตราส่วนของจุดหมายปลายทาง (Embark) และพล็อต Pie chat พร้อมทั้งให้ Queenstown เป็น explode

A= df['Ticket'].count()
R = df.groupby('Embarked')['Ticket'].count()
R2 = (R/A)*100
explode = [0,0.3,0]
plt.title('Title:Embarked')
plt.pie(R.values,labels=  [ 'Queenstown' if i == 'Q' else 'Southampton' if i== 'S' else 'Cherbourg'  for i in R.index], \
       explode=explode)
plt.show()
```

```python
#12.สร้างพล็อตแบบกระจายโดยให้อายุเป็นแกน X และ ราคาตั๋วโดยสารเป็นแกน Y
x=df['Age']
y=df['Fare']
plt.figure(figsize=[5,5])
plt.scatter(x,y,c='y')
plt.show()
```

```python
#13.สร้าง bar chart โดยใช้ข้อมูลจากราคาตั๋วโดยสารเฉลี่ยแต่ละชั้น 
x = np.sort(df['Pclass'].unique())
y = df.groupby('Pclass')['Fare'].mean()

fig = plt.figure()
ax = fig.add_axes([0,0,1,1,])
ax.set_xlabel('Tiket Class')
ax.set_ylabel('Fare Average')
ax.bar(x,y)

```

```python
#14. สร้าง bar chart โดยใช้ข้อมูลจากราคาตั๋วโดยสารต่ออายุของผู้โดยสาร
x=df['Age']
y=df['Fare']
plt.figure(figsize=[10,5])
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
plt.bar(x,y)
plt.show()
```

```python
#15. สร้าง bar chart โดยใช้ข้อมูลผู้ที่นามสกุลซ้ำกัน โดยความถี่เป็นแกน Y
LastN = df['Name'].str.split(',',expand=True)[0]
Amount = LastN.value_counts()
D = Amount[Amount>1]
x = D.index
y = D.values

plt.figure(figsize=[30,10])

plt.ylabel('Amount Duplicated Name')
plt.xlabel('Lastname')
plt.bar(x,y)
```
